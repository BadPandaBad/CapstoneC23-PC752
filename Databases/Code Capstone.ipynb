# **Generate Data**

## There will be four features and one target data that we need, including:

### 1. Heart Rate

*  Poor: RHR above 90 bpm
*  Fair: RHR between 76-89 bpm
*  Good: RHR between 61-75 bpm
*  Very good: RHR between 46-60 bpm
*  Excellent: RHR below 45 bpm

### 2. Physical Activity

*  Sedentary: Less than 5,000 steps per day. This indicates a predominantly sedentary lifestyle with minimal physical activity.
*  Low activity: 5,000 to 7,499 steps per day. This range suggests some level of physical activity but falls short of the recommended minimum.
*  Moderately active: 7,500 to 9,999 steps per day. This indicates a moderate level of physical activity, approaching or meeting the recommended guidelines.
*  Active: 10,000 to 12,499 steps per day. This range suggests an active lifestyle and meeting the recommended minimum for overall health benefits.
*  Highly active: 12,500 or more steps per day. This indicates a high level of physical activity, exceeding the recommended guidelines and contributing to enhanced fitness and well-being.

### 3. Calorie Intake
*  Poor: Consuming significantly fewer calories than the estimated individual energy needs, typically below 1,200-1,500 calories per day for most adults. This can result in inadequate nutrient intake and potential health risks.
*  Fair: Consuming calories within a moderate range but potentially below or above the estimated individual energy needs. This range can vary widely, but it generally falls between 1,500-1,800 calories per day for most adults. However, it's important to note that individual needs can vary significantly, and a fair caloric intake may not meet everyone's requirements.
*  Good: Consuming calories that align with the estimated individual energy needs and provide a balanced intake of essential nutrients. For most adults, this typically falls within the range of 1,800-2,500 calories per day, depending on factors like age, gender, activity level, and body composition.
*  Very good: Consuming calories that support individual energy needs, promote a healthy weight, and provide a well-rounded intake of essential nutrients. The specific range for a very good caloric intake can vary widely depending on individual factors. It generally falls within the range of 2,500-3,000 calories per day for most active adults.
*  Excellent: Consuming calories that optimize individual energy needs, support a healthy weight, and provide a well-balanced intake of essential nutrients. The specific range for an excellent caloric intake can vary greatly depending on individual factors and goals. It often exceeds 3,000 calories per day for active individuals with higher energy needs.

### 4. Sleep Duration
*  Poor: This category indicates a significantly inadequate amount of sleep.
*  Fair: This category suggests a moderately insufficient amount of sleep.
*  Good: This category represents a satisfactory amount of sleep.
*  Very Good: This category indicates a high-quality amount of sleep.
*  Excellent: This category represents an exceptional amount of sleep.

### 5. Physics Health
*  Poor: This category indicates a low level of physical health. Individuals classified as "Poor" may have significant physical limitations, chronic health conditions, or difficulty performing basic physical tasks.
*  Fair: This category suggests a moderate level of physical health. Individuals classified as "Fair" may have some limitations or health issues that affect their physical abilities but can still perform daily activities with some difficulty.
*  Good: This category represents a satisfactory level of physical health. Individuals classified as "Good" generally have average physical capabilities and are able to engage in most physical activities without major limitations or difficulties.
*  Very Good: This category indicates a high level of physical health. Individuals classified as "Very Good" have above-average physical capabilities and can engage in a wide range of physical activities with ease.
*  Excellent: This category represents an exceptional level of physical health. Individuals classified as "Excellent" have outstanding physical capabilities and can perform demanding physical tasks or activities at a superior level.

```python
import random
import numpy as np
import pandas as pd

num_samples = 20000
```

```python
sleep_duration_range = (4.0, 10.0)
physical_activity_range = (100, 10000)
heart_rate_range = (45, 125)
caloric_intake_range = (900, 3000)
```

```python
indicator_ranges = {
    'Poor': {'SleepDuration': (4.0, 6.0), 'PhysicalActivity': (100, 2000), 'HeartRate': (100, 125), 'CaloricIntake': (900, 1200)},
    'Fair': {'SleepDuration': (6.0, 7.0), 'PhysicalActivity': (2000, 4000), 'HeartRate': (90, 100), 'CaloricIntake': (1200, 1500)},
    'Good': {'SleepDuration': (7.0, 8.0), 'PhysicalActivity': (4000, 6000), 'HeartRate': (80, 90), 'CaloricIntake': (1500, 1800)},
    'Very Good': {'SleepDuration': (8.0, 9.0), 'PhysicalActivity': (6000, 8000), 'HeartRate': (70, 80), 'CaloricIntake': (1800, 2400)},
    'Excellent': {'SleepDuration': (9.0, 10.0), 'PhysicalActivity': (8000, 10000), 'HeartRate': (45, 70), 'CaloricIntake': (2400, 3000)}
}
```

```python
labels = ['Poor', 'Fair', 'Good', 'Very Good', 'Excellent']
```

```python
sleep_duration = []
physical_activity = []
heart_rate = []
caloric_intake = []
physics_health = []
```

```python
for _ in range(num_samples):
    label = np.random.choice(labels)
    ranges = indicator_ranges[label]
    sleep_duration.append(np.random.randint(*ranges['SleepDuration']))
    physical_activity.append(np.random.randint(*ranges['PhysicalActivity']))
    heart_rate.append(np.random.randint(*ranges['HeartRate']))
    caloric_intake.append(np.random.randint(*ranges['CaloricIntake']))
    physics_health.append(label)
```

```python
# Add noise to the attribute values
sleep_duration = np.round(np.array(sleep_duration) +- np.random.normal(0, 0.2, num_samples), 2)
physical_activity = np.round(np.array(physical_activity) +- np.random.normal(0, 300, num_samples)).astype(int)
heart_rate = np.round(np.array(heart_rate) +- np.random.normal(0, 10, num_samples), 2).astype(int)
caloric_intake = np.round(np.array(caloric_intake) +- np.random.normal(0, 200, num_samples), 2).astype(int)
```

```python
health_data = pd.DataFrame({
    'SleepDuration': sleep_duration,
    'PhysicalActivity': physical_activity,
    'HeartRate': heart_rate,
    'CaloricIntake': caloric_intake,
    'PhysicsHealth': physics_health
})
```

# **Data Preprocessing**

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from keras.callbacks import History

health_data = data.copy()
```

```python
# Split the dataset into features and labels
features = health_data.drop('PhysicsHealth', axis=1)
labels = health_data['PhysicsHealth']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, stratify=labels, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convert the labels to categorical
y_train_categorical = pd.get_dummies(y_train)
y_test_categorical = pd.get_dummies(y_test)
```

# **Modelling**

```python
# Build the TensorFlow model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(5, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
# Train the model
history = History()
History = model.fit(X_train_scaled, y_train_categorical, epochs=10, batch_size=32, verbose=1, validation_data=(X_test_scaled, y_test_categorical), callbacks=history)

# Evaluate the model on the test set
loss, accuracy = model.evaluate(X_test_scaled, y_test_categorical)
print(f'\nTest Loss: {loss:.4f}')
print(f'Test Accuracy: {accuracy:.4f}')
```

```python
plt.figure(figsize=(12,3))
plt.plot(History.history['accuracy'])
plt.plot(History.history['val_accuracy'])
plt.title('Model Accuracy', fontsize=14)
plt.ylabel('Accuracy', fontsize=12)
plt.xlabel('Epoch', fontsize=12)
plt.legend(['train','test'], loc='lower right')
plt.show()

plt.figure(figsize=(12,3))
plt.plot(History.history['loss'])
plt.plot(History.history['val_loss'])
plt.title('Model Loss', fontsize=14)
plt.ylabel('Loss', fontsize=12)
plt.xlabel('Epoch', fontsize=12)
plt.legend(['train','test'], loc='upper right')
plt.show()
```

```python
# Make predictions on new data
new_data = pd.DataFrame({
    'SleepDuration': [8.84, 7.99, 6.9, 5.86, 3.91],
    'PhysicalActivity': [8124, 7840, 5688, 3808, 1219],
    'HeartRate': [49, 72, 80, 93, 130],
    'CaloricIntake': [2313, 1931, 1670, 1656, 1466]
})
new_data_scaled = scaler.transform(new_data)
predictions = model.predict(new_data_scaled)
predicted_labels = np.argmax(predictions, axis=1)
predicted_classes = [labels[i] for i in predicted_labels]
print('Predicted Classes:', predicted_classes)
```
